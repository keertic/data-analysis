## Part A:

#### Introduction: 
<p>I wanted to explore how engine size MPG (both city and highway) were correlated and if the type of the car also helped determined the engine size and mpg. I belive that engine size and mpg should have a slight negative correlation since sports cars have larger engines and a low mpg value. I assume that this remains true for most types of cars.</p>

#### Analysis:
<p> To explore the relationship between engine size and MPG I created two scatter plot that depicts the MPG city and MPG highway along with engine size. I then grouped each point by the type of car to further understand if it affected either engine size or mpg.</p> 

<p>I found that MPG City and MPG Highway VS Engine Size both follow a negative correlation that is also slightly curvilinear. The data is what I expected with the higher the MPG the smaller the Engine Size.</p>

<p>I also thought it was interesting to see that although each type of car had different engine sizes the MPG did not change much based on the engine size. For example, if we look at SUV's on the MPG City vs Engine Size scatter plot, the engine size fluctuated from 2 to 7 while the mpg never went higher than 20. I believe this suggests that MPG is more correlated with the car type than it is with engine size. Because of this, I decided to make another scatter plot with MPG City and Type of car and found that MPG city does change slightly within the type of cars </p>

<p>I decided to observe the statistics of MPG City and MPG Highway because according to the scatter plot MPG Highway should be greater than MPG City minus a few data points where the car had lower MPG Highway than MPG City. According to the descriptive statistics I found that MPG Highway was in fact greater than MPG City</p>


```sas
title "MPG City vs Engine Size";
proc sgplot data=sashelp.cars;
scatter x=EngineSize y=mpg_city/group=Type;
run;
```

```sas
title "MPG Highway vs Engine size";
proc sgplot data=sashelp.cars;
scatter x=EngineSize y=mpg_highway/group=Type;
run;
```

```sas
title "MPG City vs Engine Size";
proc sgplot data=sashelp.cars;
scatter x=type y=mpg_city;
run;
```

```sas
title "Sashelp.car dataset:descriptive statistics";
proc univariate data=sashelp.cars;
var mpg_highway;
run;
```

```sas
title "Sashelp.car dataset:descriptive statistics";
proc univariate data=sashelp.cars;
var mpg_city;
run;
```

## Part B:

#### Introduction: 
<p>I wanted to look into what contributes to the weight of fish. If fish are heavier are they necessarily longer, or are they wider? 
</p>

#### Analysis: 
<p>
To investigate the weight of fish, I compared the weight of the fish with the length(from nose to tail) and the weight with the width of the fish. 

In the first scatter plot I compared the weight of the fish with the length of the fish while grouping by species just to see if the results were affected by species. I found that the Weight was affected my Length which followed a parabolic curve. The length goes up with weight but not in a linear manner. Most species follow this trend, except for Smelt and Pike. This data makes sense because Smelt is a very small fish while Pike is a larger fish. 
The scatter plot that compared the weight with width still shows a parabolic curve although it is more linear than the scatterplot comparing the length. This causes me to believe that weight is affected more by width instead of length. 

I then looked at the mean of the width and weight to get a better understanding of the relationship between them. Smelt is just a very small fish in both length and width, which doesn't help us understand more about the species. I think it would be useful to gather a larger dataset of small fish and do further analysis. Pike is fish that varies drastically in weight and doesn't vary that much in width which would cause us to believe that the weight of pike is more correlated with length than with width. 

Although the scatterplots may suggest that width is a larger factor in understanding the weight of the fish, after looking into the descriptive statistics, I believe that it depends on the individual species of fish. 
</p>

```sas
title "Length vs Weight";
proc sgplot data=sashelp.fish;
scatter x=Length3 y=Weight/group=Species;
run;
```

```sas
title "Width vs Weight";
proc sgplot data=sashelp.fish;
scatter x=Width y=Weight/group=Species;
run;
```

```sas
title "Descriptive Statistics: Weight, Width";
proc means data=sashelp.fish;
class Species;
var weight width;
run;
```

```sas
title "Descriptive Statistics: Weight, Length";
proc means data=sashelp.fish;
class Species;
var weight length3;
run;
```

#### Conclusion: 

<p>In this report, I discussed some of the relationships between Weight, Length, and Width of the species of fish. I was trying to determine what contributed to the weight of the fish more. In the end, I found that although it looks like width contributes to the weight more there needs to be further analysis into specific species of fish before any conclusion is made.</p>
